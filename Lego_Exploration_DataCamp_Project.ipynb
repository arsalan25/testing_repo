```
# Goal: Explore LEGO data over the years
# Remix of DataCamp project without referencing documentation
# Project: https://projects.datacamp.com/projects/10
# Data: https://rebrickable.com/downloads/

# Setup
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# How many colors are present?
colors = pd.read_csv('colors.csv')
colors.head()
```

```
print(f'The shape of the dataset is: {colors.shape}')
print(f'Meaning the total number of colors is: {colors.shape[0]}')
```

```
# How many colors have transparency?
colors.groupby('is_trans').count()
```

```
# How many LEGO sets are there per year?
sets = pd.read_csv('sets.csv')
sets.head()
```

```
sets_by_year = sets.groupby('year', as_index=False).count()
sets_by_year.head()
```

```
plt.figure(figsize=(8,8))
sns.lineplot(x='year', y='set_num', data=sets_by_year)
plt.title('Number of LEGO Sets Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Sets')
```

```
# What is the average number of set parts per year?
parts_per_year = sets.groupby(['year', 'num_parts'], as_index=False).mean()
```

```
plt.figure(figsize=(10,10))
sns.lineplot(x='year', y='num_parts', data=parts_per_year)
plt.title('Average Number of Parts per Set per Year')
plt.xlabel('Year')
plt.ylabel('Average Number of Parts per Set')
```
