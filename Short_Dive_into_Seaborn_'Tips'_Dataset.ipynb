```
# Explore Seaborn's Tips Dataset
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
```

```
data.head()
```

```
# Explore Bill : Tip Relationship
plt.figure(figsize=(10,8))
sns.scatterplot(x='total_bill', y='tip', data=data, hue='sex', palette='deep', size='size')
plt.title('How do you tip?')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
```

```
# Brief exploration
data.describe()
```

```
data.info()
# All columns have same number of values
```

```
plt.figure(figsize=(10,6))
sns.swarmplot(x='day', y='tip', data=data)
plt.title('How well do you tip?')
plt.xlabel('Day')
plt.ylabel('Tip Amount')
```
